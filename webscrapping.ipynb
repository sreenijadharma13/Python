```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import csv
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.motorsportmagazine.com/database/drivers/fernando-alonso/')
```

```python
html = driver.page_source
```

```python
table = driver.find_element(By.TAG_NAME, 'table')
```

```python
print(table.text)
```

```python
headers = []
for th in table.find_elements(By.TAG_NAME, 'th'):
    headers.append(th.text.strip())
```

```python
print(headers)
```

```python
rows = []
for tr in table.find_elements(By.TAG_NAME, 'tr'):
    row = [td.text.strip() for td in tr.find_elements(By.TAG_NAME, 'td')]
    if row:
        rows.append(row)
```

```python
print(rows)
```

```python
import csv 
csv_file = r'C:\Users\srini\Downloads\fernando-alonso.csv'
with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # Write the header if available
    if headers:
        writer.writerow(headers)
    
    # Write the rows
    writer.writerows(rows)

print(f"Data saved to {csv_file}")
```
